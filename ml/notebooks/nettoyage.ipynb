```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
```

<h4>Suppression des images corrompues,non lisibles et trop flou</h4>

```python
data_path = "../data/raw"
```

```python
classes = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
classes
```

```python
def class_repartition(classes,data_path):
    class_counts = {}
    for c in classes: 
        c_path = os.path.join(data_path, c)
        num_images = len([f for f in os.listdir(c_path) if os.path.isfile(os.path.join(c_path, f))])
        class_counts[c] = num_images
        
    return pd.DataFrame(list(class_counts.items()), columns=["Classe", "Nombre d'images"])
```

```python
#suppression des images corrompues
for i,c in enumerate(classes):
    c_path = os.path.join(data_path,c)
    lis_img = os.listdir(c_path)
    for img in lis_img:
        try:
            img_resized = Image.open(os.path.join(c_path,img))
        except Exception as e:
            os.remove(os.path.join(c_path,img))
```
