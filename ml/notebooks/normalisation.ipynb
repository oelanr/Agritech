```python
import cv2 
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
```

```python
data_raw = "../data/raw/"
```

```python
img_train = cv2.imread("../data/raw/Bacterialblight/BACTERAILBLIGHT3_004.jpg")
img_train = cv2.cvtColor(img_train,cv2.COLOR_BGR2RGB)
```

```python
img_to_gray = cv2.cvtColor(img_train,cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(13,3))
plt.subplot(1,2,1)
plt.imshow(img_train)
plt.title("Original")
plt.subplot(1,2,2)
plt.imshow(img_to_gray,cmap="gray")
plt.title("Gray")
```

```python
thresh = cv2.adaptiveThreshold(img_to_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,7)
```

```python
plt.figure(figsize=(13,3))
plt.subplot(1,3,1)
plt.imshow(img_train)
plt.title("Original")
plt.subplot(1,3,2)
plt.imshow(img_to_gray,cmap="gray")
plt.title("Gray")
plt.subplot(1,3,3)
plt.imshow(thresh,cmap="gray")
plt.title("Threshold")
```

```python
#convert [0-255] to [0-1]
thresh = thresh.astype(np.float32)
thresh /=255
```
