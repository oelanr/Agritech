```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from PIL import Image
```

```python
data_path = "../data/raw"
```

```python
classes = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
classes
```

```python
#CV2
W,H=256,256
for c in classes:
    c_path = os.path.join(data_path,c)
    list_img = os.listdir(c_path)
    for img in list_img:
        img_resized = cv2.imread(os.path.join(c_path,img))
        img_resized = cv2.resize(img_resized,(W,H))
        cv2.imwrite(os.path.join(c_path,img),img_resized)
```

```python
img_shapes = []
for c in classes: 
    c_path = os.path.join(data_path, c)
    images = os.listdir(c_path)
    for img_name in images:
        img = Image.open(os.path.join(c_path, img_name))
        img_shapes.append(img.size)

df_shapes = pd.DataFrame(img_shapes, columns=["Largeur", "Hauteur"])
```
